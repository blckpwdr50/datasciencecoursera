## **Part 1: Exploratory Data Analysis with Stars**
---
There are at least 100 billion stars in the Milky Way. The Stars dataset contains information about several features of 200+ stars. This dataset categorizes data by type of star and star color, making it useful for classification tasks. Most importantly, the dataset can be used to show that stars follow the pattern of the Hertzsprung-Russell Diagram (HR diagram), which is one of the most important tools in the study of stellar evolution and is a scatter plot that shows the relationship between temperature, luminosity, star class, and star color.

**Below is information about each of the attributes:**

- `temperature`: Absolute temperature (in K)
- `brightness`: Relative luminosity with respect to the Sun (L)
- `size`: Relative radius with respect to the Sun (R)
- `color`: Star color
- `class`: Star type (0: Brown Dwarf, 1: Red Dwarf, 2: White Dwarf, 3: Main Sequence, 4: Supergiant, 5: Hypergiant)



**Before continuing, run the cell below to load the dataset.**

```python
url = "https://raw.githubusercontent.com/the-codingschool/TRAIN/main/stars/stars.csv"
stars_df = pd.read_csv(url)
print(stars_df)
```

### **Problem #1.1**

How many stars are recorded in the dataset?

```python
stars_data= stars_df.shape
print('There are ' + str(stars_data[0]) +' stars in this table.')

```

### **Problem #1.2**

What is the average star temperature?

```python
print('The average temperature of stars in this table is '+str(stars_df['temperature'].mean()))
```

### **Problem #1.3**

What is the most common star color?

```python
print(stars_df['color'].value_counts())
print('')
print("The most common star color is "+ stars_df['color'].mode())
```

### **Problem #1.4**

How many stars have a radius above 1000?

```python

stars_df['radius>1000']=''
stars_df.loc[stars_df['size']>1000,'radius>1000']=1
stars_df.loc[stars_df['size']<=1000,'radius>1000']=0

stars_df['radius>1000'].value_counts()
print('There are '+ str(stars_df['radius>1000'].sum()) + ' stars that have a radius bigger than 1000.')
```

### **Problem #1.5**

For each type of star, how many are in the dataset?

```python
stars_df['class'].value_counts()
```

### **Problem #1.6**

Find the median for `brightness`.

```python
print('The median star brightness is ' + str(stars_df['brightness'].median()))
```

### **Problem #1.7**

What is the median for `size` for rows 223-230?

```python
stars_df.loc[222:229,'size'].median()
```

### **Problem #1.8**

What is the smallest radius for a star in the dataset?

```python
stars_df['size'].min()
```

### **Problem #1.9**

What is the largest brightness among the first 10 stars?

```python
stars_df['size'].max()
```

### **Problem #1.10**

How many stars are the color 'Blue white'?

```python
stars_df["color"].value_counts()
```

### **Problem #2.4**

```python
days = [1, 2, 3, 4, 5, 6]
bowlerA = [100, 94, 132, 171, 239, 166]
bowlerB = [147, 122, 127, 184, 205, 253]
plt.scatter(days,bowlerA,marker='D',color='purple',label='Bowler A')
plt.scatter(days,bowlerB,marker='*',color='orange',label="Bolwer B ")
plt.title("Bowling Score Over Time")

plt.xlabel('Day')
plt.ylabel("Score") # COMPLETE THIS LINE
plt.legend()
plt.show()
# plot
```

### **Problem #2.5**

Using the following data, create a scatter plot. In addition:
* Make one line dashed and one line dashed and dotted.
* Make one line green and one line pink.

```python
x = [10, 20, 30, 40, 50]
y1 = [50, 100, 150, 200, 250] # y = 5x
y2 = [85, 95, 105, 115, 125] # y = x + 75

plt.plot(x,y1,color='pink',ls='-.',label='y = 5x')
plt.plot(x,y2,color='green',ls='--',label="y = x + 75")
 # COMPLETE THIS LINE
plt.legend()
plt.show()
```

### **Problem #2.6**

Using the following data, create a bar plot. In addition:
* Make bar `Blue` blue, and make the other bars gray.
* Make the x-label `Eye Color` and in the font `monospace`
* Make the y-label `Number of Students` and in the font `monospace`
* Make the Title `Student Eye Colors` and in the font `monospace`
* Make the bar width 0.5

```python
eye_color = ["Brown", "Hazel", "Blue", "Green"]
bars = [20, 9, 12, 3]

plt.bar(eye_color,bars,width=0.5,color=["gray", "gray", "blue", "gray"])

plt.title("Student Eye Color",font='monospace')

plt.xlabel('Eye Color',font='monospace')
plt.ylabel("Number of Students",font='monospace')
plt.legend()
plt.show()
```
